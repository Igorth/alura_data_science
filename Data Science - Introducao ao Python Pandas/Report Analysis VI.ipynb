# Report Analysis VI

## Create new variables

```python
import pandas as pd
```

```python
df = pd.read_csv('dados/aluguel_residential.csv', sep=';')
```

```python
df.head()
```

### Criar uma nova coluna

```python
df['Valor Bruto'] = df['Valor'] + df['Condominio'] + df['IPTU']
```

```python
df.head()
```

```python
df['Valor por metro quadrado'] = (df['Valor'] / df['Area']).round(2)
```

```python
df.head()
```

```python
df['Valor Bruto m2'] = (df['Valor Bruto'] / df['Area']).round(2)
```

```python
df.head()
```

```python
casa = ['Casa', 'Casa de Condominio', 'Casa de Vila']
```

```python
df['Tipo Agregado'] = df['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')
```

```python
df.head()
```

## Excluindo variáveis

```python
df_aux = pd.DataFrame(df[['Tipo Agregado', 'Valor por metro quadrado', 'Valor Bruto', 'Valor Bruto m2']])
```

```python
df_aux.head()
```

```python
del df_aux['Valor Bruto']
```

```python
df_aux
```

```python
df.drop(['Valor Bruto', 'Valor Bruto m2'], axis = 1, inplace = True)
```

```python
df.head()
```

```python
df.to_csv('dados/aluguel_residential.csv', sep=';', index = False)
```
